```python
import random
import pickle
import json
import numpy as np
import pickle
import nltk
from nltk.stem import WordNetLemmatizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.optimizers import SGD
import nltk
nltk.download('punkt')
nltk.download('wordnet')
from nltk import word_tokenize,sent_tokenize
import joblib
```

```python
Intents={
'ON' : [
    "Could you please turn on the LED for me?",
    "Is it possible for you to activate the LED?",
    "Would you mind switching on the LED?",
    "Can you enable the LED, please?",
    "I'd appreciate it if you could turn on the LED.",
    "If you don't mind, could you light up the LED?",
    "Is there a chance you could activate the LED?",
    "I'm wondering if you could turn on the LED for me.",
    "Might I request you to switch on the LED?",
    "Could you perform the action of turning on the LED?",
    "I'm in need of the LED being turned on, could you do that?",
    "If it's not too much trouble, could you enable the LED?",
    "Turning on the LED would be great, could you do it?",
    "Can you make the LED active for me, please?",
    "Would you be able to turn on the LED at this moment?",
    "Is turning on the LED within your capabilities right now?",
    "Can you execute the task of turning on the LED?",
    "Could you carry out the action of enabling the LED?",
    "I'm kindly requesting you to turn on the LED.",
    "Is it feasible for you to activate the LED for me?"
]
,
'OFF' : [
    "Could you please deactivate the LED?",
    "Would you mind switching off the LED?",
    "Is it possible for you to turn off the LED?",
    "Can you disable the LED for me?",
    "I'd appreciate it if you could power down the LED.",
    "Might you turn off the LED?",
    "Do you have the ability to deactivate the LED?",
    "I request the LED to be turned off, could you?",
    "If it's not too much trouble, switch off the LED.",
    "Turning off the LED would be great, can you do it?",
    "Could you perform the action of turning off the LED?",
    "I'm asking you to power down the LED, can you comply?",
    "Is it within your capabilities to deactivate the LED?",
    "Can you execute the command to turn off the LED?",
    "I'm in need of the LED being turned off, could you do that?",
    "Would you be able to switch off the LED for me?",
    "It would be appreciated if you could turn off the LED.",
    "Can you make sure the LED is turned off?",
    "Could you kindly turn off the LED?",
    "If you can, please turn off the LED for me."
]
}
```

```python
lemmatizer=WordNetLemmatizer()
Words=[]
```

```python
ignored=['?','!','.',',',"'d","'s"]
def Cleaner(Sentence):
  words=nltk.word_tokenize(Sentence)
  return [str.lower(lemmatizer.lemmatize(word)) for word in words if word not in ignored]
```

```python
for _ in Intents:
  for i in Intents[_]:
    Words.extend(Cleaner(i))
Words=sorted(list(set(Words)))
print(list(Words))
```

```python
Overall=[]
for _ in Intents:
  for i in Intents[_]:
    Tempo=[]
    temp=nltk.word_tokenize(i)
    for __ in Words:
        if __ in temp:
          Tempo.append(1)
        else:
          Tempo.append(0)
    val=[1 if _=="ON" else 0]
    Overall.append([Tempo,val])
random.shuffle(Overall)
for _ in Overall:
  print(_)
```

```python
Overall=np.array(Overall)
train_x=list(Overall[:,0])
train_y=list(Overall[:,1])
print(train_x,train_y)
model=Sequential()
model.add(Dense(128,input_shape=(len(train_x[0]),),activation="relu"))
model.add(Dense(units=1, activation='sigmoid'))
# Compiling the model with binary crossentropy loss and Adam optimizer
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
model.fit(np.array(train_x),np.array(train_y),epochs=100,batch_size=5,verbose=1)
import tensorflow as tf
my_model = model
saved_model_path = "Modelio"
my_model.save(saved_model_path)
```

```python
def MakePredictions(sent):
  Tempo=[]
  temp=nltk.word_tokenize(sent)
  for __ in Words:
      if __ in temp:
        Tempo.append(1)
      else:
        Tempo.append(0)
  c=list(model.predict([Tempo]))
  return c[0]
```
